```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_df=pd.read_csv('/content/train (2).csv')
test_df=pd.read_csv('/content/test (2).csv')
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
train_df.isnull().sum().sort_values(ascending=False)
```

```python
train_df[['Deck','CabinNum','Side']] = train_df['Cabin'].str.split('/', expand=True)
test_df[['Deck','CabinNum','Side']] = test_df['Cabin'].str.split('/', expand=True)
```

```python
train_df.drop('Cabin', axis=1, inplace=True)
test_df.drop('Cabin', axis=1, inplace=True)
```

```python
expenses=['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
```

```python
train_df['TotalSpend']=train_df[expenses].sum(axis=1)
test_df['TotalSpend']=test_df[expenses].sum(axis=1)
```

```python
train_df.drop(['Name'],axis=1,inplace=True)
test_df.drop(['Name'],axis=1,inplace=True)
```

```python
X=train_df.drop(['Transported'],axis=1)
y=train_df['Transported']
```

```python
categorical_cols=X.select_dtypes(include='object').columns
numerical_cols=X.select_dtypes(exclude='object').columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
numeric_transformer=Pipeline([
    ('imputer',SimpleImputer(strategy='median'))
])
categorical_transformer=Pipeline([
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor=ColumnTransformer([
    ('num',numeric_transformer,numerical_cols),
    ('cat',categorical_transformer,categorical_cols)
])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=300,random_state=42)
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([
    ('preprocessor',preprocessor),
    ('model',model)
])
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(
    clf,
    X,
    y,
    cv=5,
    scoring='accuracy'
)

print(scores.mean())
```
